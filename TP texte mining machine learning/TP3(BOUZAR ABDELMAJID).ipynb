## Nous allons utiliser le module urllib2 pour explorer n'importe quelle  page Web

```python
import nltk
import urllib.request
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/IPhone')
html = response.read()
print (html)
```

## Nous utilisons BeautifulSoup pour extraire le texte d'un site web format html

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'html.parser')
text = soup.get_text(strip=True)
print (text)
```

```python
tokens= nltk.word_tokenize(text)
```

```python
freq = nltk.FreqDist(tokens) 
freq.plot(20)
```

```python
from nltk.corpus import stopwords 

clean_tokens = tokens 
sr = stopwords.words('english') 
for token in tokens: 
    if token in sr: 
        clean_tokens.remove(token) 
```

## Question 2: 

Ecrire une fonction qui prend en argument le dictionnaire de frquence et qui retourne deux valeurs qui reprsentent le mot avec sa maximum frquence (mot, Maximum_frequence).

```python
#Ecriver votre code ici
def getMaxFreq(dictionary):
   
    maxkey = ""
    maxlen = 0
    for k in dictionary.keys():
        if  dictionary[k] > maxlen:
            maxlen = dictionary[k]
            maxkey = k
    return maxlen, maxkey



```

```python
getMaxFreq(freq2)
```
